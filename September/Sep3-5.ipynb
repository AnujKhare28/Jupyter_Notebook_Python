```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('heart.csv')
```

```python
df['chol'].hist()
```

```python
cate_val = []
cont_val = []
```

```python
for column in df.columns: 
    if df[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
df.hist(cont_val, figsize=(15,8))
plt.tight_layout()
plt.show()
```
