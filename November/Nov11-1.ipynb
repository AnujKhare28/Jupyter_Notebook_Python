```python
def oddEvenList(self, head):
      
        if head is None or head.next is None: 
            return head

        temp = head
        value = []

        while temp:
            value.append(temp.val)
            temp = temp.next
            if temp:
                temp = temp.next

        t = head.next
        while t:
            value.append(t.val)
            t = t.next
            if t:
                t = t.next

        tp = head
        index = 0
        while tp:
            tp.val = value[index]
            index += 1
            tp = tp.next
        
        return head
```
