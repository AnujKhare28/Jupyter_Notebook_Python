```python
def detectCycle(self, head):
    slow = head
    fast = head
    while fast is not None and fast.next is not None: 
        slow = slow.next
        fast = fast.next.next
        if slow == fast: 
            slow = head 
            while slow != fast:
                slow = slow.next
                fast = fast.next
            return slow
    return None
```
