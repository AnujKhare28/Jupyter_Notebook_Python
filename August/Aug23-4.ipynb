```python
df = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
df.groupby('Year')['Rating'].mean().sort_values(ascending=False)
```

```python
def rating(rating): 
    if rating >= 7.0: 
        return "Excellenet"
    elif rating >= 6.0: 
        return "Good"
    else: 
        return "Average"
```

```python
df['Rating_Cat'] = df['Rating'].apply(rating)
```
