```python
import pandas as pd
```

```python
dict1 = {'Name' : ['Priyang', 'Aadhya', 'Krishna','Vedant','Parshv',
                   'Mittal', 'Swarali'],
         'Marks' : [98,89,87,99,90,83,99],
         'Gender' : ['Male','Female','Female','Male','Male',
                     'Female','Female']
        }

df1 = pd.DataFrame(dict1)
df1
```

```python
def marks(x):
    return x//2
```

```python
df1['Marks'].apply(marks)
```

```python
df1['Half Marks'] = df1['Marks'].apply(marks)
```
