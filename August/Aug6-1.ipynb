# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]

plt.xlabel('languages')
plt.ylabel('No.')
plt.title('Anuj Khare')
plt.bar(x,y, width= 0.5)
plt.show()
```

```python
c = ['r','b','g','m']
plt.bar(x,y, width= 0.5, color = c, align='edge', edgecolor='y', linewidth=5)
plt.show()
```
