```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
df['Year'].value_counts()
```

```python
sns.countplot(x='Year',data=df)
plt.title('Number of Movies Per Year')
plt.show()
```

```python
df[df['Revenue (Millions)'].max() == df['Revenue (Millions)']]['Title']
```

```python
top10Rate = df.nlargest(10,'Rating')[['Title','Rating','Director']]
```

```python
top10Rate
```

```python
sns.barplot(x='Rating', y='Title', data=top10Rate, hue='Director')
plt.legend(bbox_to_anchor=(1.05,1), loc=2)
plt.show()
```

```python
df.nlargest(10, 'Revenue (Millions)')['Title']
```

```python
top10M = df.nlargest(10, 'Revenue (Millions)')[['Title','Revenue (Millions)']].\
set_index('Title')
```

```python
top10M
```

```python
sns.barplot(x='Revenue (Millions)', y=top10M.index, data=top10M)
plt.title('Top 10 Highest Revenue Movie Titles')
plt.show()
```
