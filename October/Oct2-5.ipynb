```python
import pandas as pd 
```

```python
d = {'Name':['Priyang Bhatt', 'Rashmin Panchal', 'Anil Rana', 'Rahul Patel'],
     'City': [' New York ', ' Los Angeles  ', '  Chicago', ' Houston'],
     'State': ['NY', 'CA', 'IL', 'TX']}
df = pd.DataFrame(d)
```

```python
df
```

```python
# str.lower() and str.upper()
df['Name_lower'] = df['Name'].str.lower()
df
```

```python
df['Name_upper'] = df['Name'].str.upper()
df
```

```python
# str.len()
df['Name_len'] = df['Name'].str.len()
df
```

```python
# str.strip()
df['City']
```

```python
df['City_strip'] = df['City'].str.strip()
df['City_len_old'] = df['City'].str.len()
df['City_len_new'] = df['City_strip'].str.len()
df
```

```python
# str.split()
df[['First_name', 'Last_name']] = df['Name'].str.split(' ', expand=True)
df
```

```python
# str.contains()
df[df['Name'].str.contains('Patel', case=False)]
```

```python
# str.replac()
df['State']
```

```python
df['State_new'] = df['State'].str.replace('NY', 'New York').str.replace('CA', 'Callifornia').str.replace("IL", 'Ilinois').str.replace("TX", 'Texas')
df
```

```python
# str.startswith() and str.endswith()
df[df['First_name'].str.startswith("A")]
```

```python
df[df['Last_name'].str.endswith('l')]
```

```python
# str.cat() (concatenate)
df['Full name'] = df['First_name'].str.cat(df['Last_name'], sep='-')
df
```
